# 12. Wrapup

```python
import matplotlib.pyplot as plt
```

Metrics for all models will be compared side-by-side.

```python
# Logistic regression metrics
logreg_accuracy = 0.89
logreg_precision = 0.96
logreg_recall = 0.80
```

```python
# Neural Networks metrics
nn_accuracy = 0.927195
nn_precision = 0.894734
nn_recall = 0.968313
```

```python
# Decision tree classifier metrics:
dt_accuracy = 0.927931
dt_precision = 0.898752
dt_recall = 0.964520
```

```python
# KNN classifier metrics:
knn_accuracy = 0.926232955372596
knn_precision = 0.8916558909647597
knn_recall = 0.970375103116963
```

```python
# Set the labels for the models
models = ['Logistic regression', 'NN', 'Decision Tree', 'KNN']

# Set the metrics for each model
accuracy = [logreg_accuracy, nn_accuracy, dt_accuracy, knn_accuracy]
precision = [logreg_precision, nn_precision, dt_precision, knn_precision]
recall = [logreg_recall, nn_recall, dt_recall, knn_recall]
```

```python
# Set the width of each bar
bar_width = 0.2

# Set the positions of the bars on the x-axis
r1 = range(len(models))
r2 = [x + bar_width for x in r1]
r3 = [x + bar_width for x in r2]

# Plot the bars
plt.bar(r1, accuracy, color='orange', width=bar_width, edgecolor='black', label='Accuracy')
plt.bar(r2, precision, color='grey', width=bar_width, edgecolor='black', label='Precision')
plt.bar(r3, recall, color='navy', width=bar_width, edgecolor='black', label='Recall')

# Add a horizontal line passing through the top value of each metric
plt.axhline(y=max(accuracy), color='orange', linestyle='--', label='Max Accuracy')
plt.axhline(y=max(precision), color='grey', linestyle=':', label='Max Precision', linewidth=0.5)
plt.axhline(y=max(recall), color='navy', linestyle=':', label='Max Accuracy', linewidth=0.5)

plt.ylim (0.79, 1)

# Set the x-axis tick labels
plt.xticks([r + bar_width for r in range(len(models))], models)

# Set the y-axis label
plt.ylabel('Score')

# Set the chart title
plt.title('Model Performance Metrics')

# Move the legend outside the plot and position it to the right
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

# Show the plot
plt.show()
```

Overall, the most suitable model would be **Decision Tree Classifier**.
1. It has the highest accuracy.
2. It has a decent recall.
3. Its precision is comparable to the 3 other methods but lower than that of logistic regression.
4. It is very interpretable.  
  
Due to its lower precision score, the decision tree model will produce more false positives compared to predictions using a the logistic regression model.

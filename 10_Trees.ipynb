# 10. Decision Trees

While neural networks outperformed logistic regression in terms of their prediction capabilities, the model is not interpretable. In this context, it can be used successfully to predict overall occurrence of the disease and make projections for the future in order to organise logistics and planning such as supply, infrastructure and staff requirements. In terms of a physician preventively contacting a patient to discuss the possibility of them developing prediabetes, however, would not be appropriate since it very difficult to extract information from the deep learning model that provides any explanation or rational behind the decision it made. This is the reason why one of the most interpretable models will be built and its performance on test data assessed - the Decision Tree Classifier

## Prep

### Import modules

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import joblib
from sklearn.metrics import roc_curve, roc_auc_score, precision_recall_fscore_support
from sklearn import tree
```

### Load training data

```python
X_train = pd.read_csv ('X_train_NN_4.csv', index_col = 0)
y_train = pd.read_csv ('y_train_4.csv', index_col = 0)

y_train = y_train.prediabetes_bin_y
```

## Optimize Hyperparameters For Decision Tree Classifier

### First run for hyperparameters

```python
# Define your pipeline steps
# Create the pipeline
pipeline = Pipeline([
    ('classifier', DecisionTreeClassifier())
])

# Define the hyperparameters you want to tune
hyperparameters = {
    'classifier__criterion': ['gini', 'entropy'],
    'classifier__max_depth': [None, 5, 25, 125],
    'classifier__min_samples_split': [20, 200, 20000, 200000],
    'classifier__min_samples_leaf': [10, 100, 10000, 100000],
    'classifier__max_features': ['sqrt', 'log2', 20]
}
```

```python
# Perform grid search using cross-validation
grid_search = GridSearchCV(pipeline, hyperparameters, cv=5)
grid_search.fit(X_train, y_train)
```

```python
# Retrieve the best model and its parameters
best_model = grid_search.best_estimator_
best_params = grid_search.best_params_
best_score = grid_search.best_score_

print (best_params)
print ('Validation accuracy score:', best_score)
```

The best hyperparaneters for the decision tree classifier from the set provided are displayed above. Even at this stage of optimisation, the decision tree classifier performs better than the logistic regression model with cross-validation accuracy (in lieu of test accuracy here) of 0.925 vs 0.89 for logistic regression.  
The parameters optimisation will be further refined.

### Refine hyperparameters (1)

```python
# Define the hyperparameters you want to tune
hyperparameters = {
    'classifier__criterion': ['gini'],
    'classifier__max_depth': [70, 125, 135],
    'classifier__min_samples_split': [20, 40, 60],
    'classifier__min_samples_leaf': [97, 100, 105],
    'classifier__max_features': [18, 20, 22]
}
```

```python
# Perform grid search using cross-validation
grid_search = GridSearchCV(pipeline, hyperparameters, cv=5)
grid_search.fit(X_train, y_train)
```

```python
# retrieve the best hyperparameters
# Retrieve the best model and its parameters
best_model1 = grid_search.best_estimator_
best_params1 = grid_search.best_params_
best_score1 = grid_search.best_score_

print (best_params1)
print ('Validation accuracy score:', best_score1)
```

### Refine hyperparameters (2)

```python
# Define the hyperparameters you want to tune
hyperparameters = {
    'classifier__criterion': ['gini'],
    'classifier__max_depth': [135, 140, None],
    'classifier__min_samples_split': [60, 120, 700],
    'classifier__min_samples_leaf': [83, 91, 97],
    'classifier__max_features': [22, 25, 147]
}
# Perform grid search using cross-validation
grid_search = GridSearchCV(pipeline, hyperparameters, cv=5)
grid_search.fit(X_train, y_train)

# Retrieve the best model and its parameters
best_model2 = grid_search.best_estimator_
best_params2 = grid_search.best_params_
best_score2 = grid_search.best_score_

print (best_params2)
print ('Validation accuracy score:', best_score2)
```

The decision tree classifier appears to perform quite well. A classifier will be trained on the entirety of the training dataset and its performance assessed on the test dataset to compare to the other models.

## Build a Decision Tree Classifier and Assess Performance

```python
# build classifier with the best parameters

# Remove 'classifier__' prefix from dictionary keys
params = {key.replace('classifier__', ''): value for key, value in best_params2.items()}

# Create the decision tree classifier with the modified parameters
best_DT = DecisionTreeClassifier(**params)

# Fit the classifier on your training data
best_DT.fit(X_train, y_train)

# Save model:
joblib.dump (best_DT, 'best_DT_model.pkl')
```

```python
# Load test data:
X_test = pd.read_csv ('X_test_NN_4.csv', index_col = 0)
y_test = pd.read_csv ('y_test_4.csv', index_col = 0)

y_test = y_test.prediabetes_bin_y
```

```python
# Get the predicted probabilities for the positive class
y_pred_proba = best_DT.predict_proba(X_test)[:, 1]
y_pred = best_DT.predict (X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
```

The optimised decision tree classifier outperforms both logistic regression and neural network model:
- accuracy : 0.9279 > 0.9272
- precision : 0.8988 > 0.8947
- recall : 0.9645 < 0.968313
- f1 score : 0.9305 > 0.930071  
  
One of the most important advantages of the decision tree classifier, however, is its interpretability allowing for:
- interaction between physician and patient based on records and model
- extraction of novel insights about future research investigating the association between a specific feature and the condition
- usage for making predictions and projections (like for the neural networks model) since the decision tree classifier performs very well.  
  
So far, the clear winner is the **DECISION TREE CLASSIFIER** based on its properties and performance.

#### More performance metrics

```python
# Get the predicted probabilities for the positive class
y_pred_proba = best_DT.predict_proba(X_test)[:, 1]

# Compute the false positive rate, true positive rate, and threshold values
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)

# Compute the AUC (Area Under the Curve)
auc = roc_auc_score(y_test, y_pred_proba)

# Plot the ROC curve
plt.plot(fpr, tpr, label='ROC Curve (AUC = {:.2f})'.format(auc))
plt.plot([0, 1], [0, 1], 'k--', label='Random')
plt.text(0.6, 0.3, ('AUC = ' + str(round(auc, 4))), fontsize = 16)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend()
plt.show()
```

```python
# Check accuracy for different thresholds
# Define threshold cut-offs:
thresholds = np.linspace(0.1, 0.97, 9)
thresholds = np.append (thresholds, 0.5)
thresholds = np.sort (thresholds)
```

```python
# Define lists to hold metrics:
accuracies = []
precisions = []
recalls = []
f1s = []

# Obtain model metrics for each decision threshold
for v in thresholds:
    print (v)
    predictions = (y_pred_proba > v).astype(int)
    # Obtain metrics
    accuracies.append ( accuracy_score(y_test, predictions) )
    precisions.append ( precision_score(y_test, predictions) )
    recalls.append ( recall_score(y_test, predictions) )
    f1s.append ( f1_score(y_test, predictions) )

# Print dataframe with metrics corresponding to each threshold
metrics_thresholds = pd.DataFrame (
    {
        'threshold' : thresholds,
        'accuracy' : accuracies,
        'precision' : precisions,
        'recall' : recalls,
        'f1 score' : f1s
    }
)
display (metrics_thresholds)
```

```python
# Plot results and indicate 0.5 threshold and metrics with best precision and recall (0.535)
    
# Plot train and test metrics in 3 separate panels
# Plotting the results
fig, axes = plt.subplots(2, 2, figsize=(15, 9))

# Plot accuracy
axes[0, 0].plot(thresholds,
             accuracies,
             marker = 'o',
             color = 'red'
            )
axes[0, 0].axvline (x = 0.5,                  
                 color = 'orange',
                 linestyle = ':')
axes[0, 0].axvline (x = 0.535, 
                 color = 'blue',
                 linestyle = ':')
axes[0, 0].set_xlabel ('P-value threshold')
axes[0, 0].set_ylabel ('Accuracy')
axes[0, 0].set_title ('Accuracy as a function of prob. threshold')

# Plot precision
axes[0, 1].plot(thresholds,
             precisions,
             marker = 6,
             linestyle = '--',
             color = 'red'
            )
axes[0, 1].axvline (x = 0.5,                  
                 color = 'orange',
                 linestyle = ':')
axes[0, 1].axvline (x = 0.535, 
                 color = 'blue',
                 linestyle = ':')
axes[0, 1].set_xlabel ('P-value threshold')
axes[0, 1].set_ylabel ('Precision')
axes[0, 1].set_title ('Precision as a function of prob. threshold')

# Plot recall
axes[1, 0].plot(thresholds,
             recalls,
             marker = 7,
             linestyle = ':',
             color = 'red'
            )
axes[1, 0].axvline (x = 0.5,                  
                 color = 'orange',
                 linestyle = ':')
axes[1, 0].axvline (x = 0.535, 
                 color = 'blue',
                 linestyle = ':')
axes[1, 0].set_xlabel ('P-value threshold')
axes[1, 0].set_ylabel ('Recall')
axes[1, 0].set_title ('Recall as a function of prob. threshold')

# Plot f1 score
axes[1, 1].plot(thresholds,
             f1s,
             marker = 's',
             linestyle = '-.',
             color = 'red'
            )
axes[1, 1].axvline (x = 0.5,                 
                 color = 'orange',
                 linestyle = ':',
                   label = 'prob. threshold: 0.500')
axes[1, 1].axvline (x = 0.535, 
                 color = 'blue',
                 linestyle = ':',
                    label = 'prob. threshold: 0.535'
                   )
axes[1, 1].set_xlabel ('P-value threshold')
axes[1, 1].set_ylabel ('f1 score')
axes[1, 1].set_title ('f1 score as a function of prob. threshold')

# place legend outside
last_plot_ax = axes[1, 1]
last_plot_ax.legend(bbox_to_anchor = (1.0, 1), loc = 'upper left')

plt.tight_layout()

plt.show()
```

The best probability threshold (striking a balance between precision and recall) is clearly at 0.5. At this value, there is a slight decrease in precision compared to a threshold of 0.535, but the f1 score (balance between precision and recall) indicates that overall a threshold of 0.5 performs best.

## Model interpretability

As noted previously, the Decision Tree Classifier is a very interpretable model. One can just follow the tree branches to arrive at the final prediction. The illustration below shows the whole model.

```python
fig, ax = plt.subplots(figsize=(12, 8))
tree.plot_tree(best_DT,
               class_names = ['non-prediabetes', 'prediabetes'],
               feature_names = X_train.columns,
               rounded = True,
               impurity = False,
               filled = True, ax=ax)
plt.show()
```

The colour indicates the certainty the leaf is prediabetic (blue) or non-prediabetic (orange). This measure is based on how homogeneous that leaf/branc is: the more homogeneous (measurements of the same class) the more certainty one has about the prediction corresponding to the trajectory on the tree.  
And here are the top few branches.

```python
fig, ax = plt.subplots(figsize=(12, 8))
tree.plot_tree(best_DT,
               class_names = ['non-prediabetes', 'prediabetes'],
               feature_names = X_train.columns,
               rounded = True,
               impurity = False,
               filled = True,
               max_depth = 2, ax=ax)
# Enlarge the font size of each box
for text in ax.texts:
    text.set_fontsize(12)

plt.show()
```

Just like in the case of logistic regression, the several most important predictors (top branches) appear to be calcium levels, body mass index and triglycerides. For instance, patients whose calcium levels are below 5.45 (top node) and whose body mass index is below 113.695 are all non-prediabetic with very high certainty.

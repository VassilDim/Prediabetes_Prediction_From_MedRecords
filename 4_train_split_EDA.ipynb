# 4. Test/Train Split and Exploratory Data Analysis

**RECAP:** From **3. Consolidate Data** a single consolidated data containing medical record information (columns) for each patient (rows) is available. Moreover, all values within this consolidated table are numeric allowing for statistical and machine learning models to be built and test.  
The current section will perform test/train split in order to prevent data leakage and use the train split for some exploratory analyses of the data.

## Prep

### Import modules

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import umap
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

### Load data

```python
# Load data:
df = pd.read_csv ('data_for_model.csv', index_col = 0)
```

Verify and make sure only a single column indicating prediabetic status present in consolidated table.

```python
# Check for prediabetes column
columns_with_prediabetes_bin = [col for col in df.columns if 'prediabetes_bin' in col]
print(columns_with_prediabetes_bin)

# Delete column prediabetes_bin_x
df.drop (columns = ['prediabetes_bin_x'], inplace = True)
```

## Obtain feature and target variables

The consolidated table will be divided in target `y` indicating the prediabetic status of each patient and feature table `X` containing the numericalised data extracted from the  medical records for each patient.

```python
# Extract target variable
y = df.prediabetes_bin_y

# Extract features table
X = df.iloc[:, :-1]

# Clean up to liberate memory
del df

# Save X
X.to_csv('X.csv')
# Save y
y.to_csv('y.csv')

print(f'Dimensions of the features table are {X.shape[0]} by {X.shape[1]}')
```

## Target variable distribution

Load the target variable, if necessary.

```python
# y = pd.read_csv ('y.csv', index_col = 0)
# y = y.prediabetes_bin_y
```

Obtain counts for prediabetic (1) and non-prediabetic patients (0).

```python
# obtain values for each category:
y4hist = y.value_counts()
```

Plot results to ascertain the classes for the target feature are well balanced.

```python
# plot values
plt.figure()
plt.bar(y4hist.index, y4hist.values)
# Adding labels to each bar
for i, v in enumerate(y4hist.values):
    plt.text(i, v, str(v), ha='center', va='top', fontsize = 18, color = 'white')
# Changing the x-axis ticks
plt.xticks(y4hist.index, ['non-prediabetic', 'prediabetic'], fontsize = 14)
plt.ylabel ('Patient count', fontsize = 14)
plt.title ('Prediabetic vs non-prediabetic patient count',
           fontsize = 16)
plt.show()
```

Perfectly balanced target categories with a little over 400,000 entries for each class. Remember that the balancing act was achieved by downsampling the non-prediabetic class in **3. Consolidate Data**.

```python
# Clean up:
del y4hist
```

## Train / Test Split

In order to prevent leakage, the data will be split into a training set (70%) and testing set (30%). The split is stratified to the target variable so that test and train are split according to the distribution of the target variable categories, which is 50/50.

```python
# Split to get 30% of data for testing and the rest for training
X_train, X_test, y_train, y_test = train_test_split (X, 
                                                     y, 
                                                     test_size = 0.3, 
                                                     stratify = y)

# Save test and train sets
X_train.to_csv ('X_train.csv')
X_test.to_csv ('X_test.csv')
y_train.to_csv ('y_train.csv')
y_test.to_csv ('y_test.csv')
```

```python
# Clear everything but the train set to save memory
del X, y, X_test, y_test
```

## Dimensionality reduction and visualisation

In this section, a small subset of the training data (10%) will be obtained and plotted in 2 dimensions after dimensionality reduction in order to assess whether there is any obvious clusters that can be observed at te macro level. Note that due to the small sample size, it may not be possible to visualise any clustering.

The complete training data will be loaded first. It will be scaled and then a small subset of 10% will be obtained for dimensionality reduction and data visualisation.

```python
# Read training data containing independent variables
X_train = pd.read_csv ('X_train.csv', index_col = 0)
```

### Scaling and obtaining 10% subsample

The best scaling here would be MinMax-scaling as many of the vairables are binary (0/1) and most of the rest can be qualified as counts as they are non-negative integers. Overall, many 0 values are present (coinciding with minimum). Therefore scaling based on the Gaussian distribution here would not be appropriate. Nevertheless, the effects of such scaling will also be examined.  
  
Some of the models that I plan to fit require scaled data as they are distance-based (KNN and logistic regression) whereas others are generally less sensitive to scaling (decision trees, random forests and neural networks). Importantly, however, dimensionality reduction is very sensitive to distances and thus it will be performed only on scaled data.
  
Selecting the 10% subset is analogous to train-test split and will be stratified based on the response variable.

```python
# Scale data
X_train_scaled = StandardScaler().fit_transform(X_train)
X_train_minmax = MinMaxScaler().fit_transform(X_train)

# clean up to liberate memory
del X_train

# Response variable required for subsetting (stratifying)
y_train = pd.read_csv ('y_train.csv', index_col = 0)
y_train = y_train['prediabetes_bin_y'].squeeze()

# Subset standard-scaled
X_train_10_scaled, _, y_train_10, _ = train_test_split (X_train_scaled, 
                                                 y_train, 
                                                 test_size = 0.9, 
                                                 stratify = y_train)
# Subset minmax-scaled
X_train_10_minmax, _, y_train_10, _ = train_test_split (X_train_minmax, 
                                                 y_train, 
                                                 test_size = 0.9, 
                                                 stratify = y_train)

# clean up to liberate memory
del y_train, X_train_minmax, X_train_scaled
```

Data is properly formatted and of a permissive size to run t-SNE first on the 10% subset of the features table `X_train` on the local machine.

### T-SNE

```python
# Perform t-SNE dimensionality reduction
# Scaled:
X_tsne_scaled = TSNE(n_components = 2).fit_transform(X_train_10_scaled)
# MinMax:
X_tsne_minmax = TSNE(n_components = 2).fit_transform(X_train_10_minmax)
```

```python
# Create a 2x2 subplot grid
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))

# Plot the first subplot t-SNE standard scled
axes[0].scatter(X_tsne_scaled[:, 0], 
                   X_tsne_scaled[:, 1],
                   c = y_train_10 # colour datapoints based on response varaible
                  )
axes[0].set_xlabel('tSNE 1')
axes[0].set_ylabel('tSNE 2')
axes[0].set_title('tSNE plot for 10% of scaled features')

# Plot the first subplot t-SNE minMax scled
axes[1].scatter(X_tsne_minmax[:, 0], 
                   X_tsne_minmax[:, 1],
                   c = y_train_10 # colour datapoints based on response varaible
                  )
axes[1].set_xlabel('tSNE 1')
axes[1].set_ylabel('tSNE 2')
axes[1].set_title('tSNE plot for 10% of min-max-scaled features')

plt.show()
```

### UMAP

Note that UMAP does not work with standard-scaled data in our case. Therefore, only UMAP for MinMax-scaled data will be performed

```python
# Apply UMAP on minmax-scaled
embedding_minmax = umap.UMAP(n_components = 2).fit_transform(X_train_10_minmax)
```

```python
plt.figure()

# Plot the first subplot UMAP minMax scled
plt.scatter(embedding_minmax[:, 0], 
                   embedding_minmax[:, 1],
                   c = y_train_10
                  )
plt.xlabel('UMAP 1')
plt.ylabel('UMAP 2')
plt.title('UMAP plot for 10% of min-max-scaled features')

plt.show()
```

### PCA

```python
# Perform PCA on standard-scaled:
pca_data_scaled = PCA(n_components = 2).fit_transform(X_train_10_scaled)
# Perform PCA on minmax-scaled:
pca_data_minmax = PCA(n_components = 2).fit_transform(X_train_10_minmax)
```

```python
# Create a 2x2 subplot grid
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20, 10))

# Plot the first subplot t-SNE standard scled
axes[0].scatter(pca_data_scaled[:, 0], 
                   pca_data_scaled[:, 1],
                   c = y_train_10
                  )
axes[0].set_xlabel('PC 1')
axes[0].set_ylabel('PC 2')
axes[0].set_title('PCA plot for 10% of scaled features')

# Plot the first subplot t-SNE minMax scled
axes[1].scatter(pca_data_minmax[:, 0], 
                   pca_data_minmax[:, 1],
                   c = y_train_10
                  )
axes[1].set_xlabel('PC 1')
axes[1].set_ylabel('PC 2')
axes[1].set_title('PCA plot for 10% of min-max-scaled features')

plt.show()
```

# K-Nearest Neighbors (KNN) Classifier

The models that have been built and trained so far belong to 2 classes:
- parametric
    - logistic regression
    - neural networks
- non-parametric
    - decision trees  
  
  
While decision trees are considered non-parametric classifiers, they are relatively unflexible. Therefore I have decided to optimise and test a more flexible non-parametric model - KNN classifier.  
  
Another important reason to employ a KNN classifier is that, unlike decision trees, it is very robust to data points that are different from the original training set (completely different readings that were never encountered during training).

## Prep

### Import modules

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report
from sklearn.metrics import roc_curve, roc_auc_score, precision_recall_fscore_support
from sklearn.metrics import precision_recall_curve, average_precision_score
from sklearn.metrics import confusion_matrix
import seaborn as sns
```

### Load training data

```python
X_train = pd.read_csv ('X_train_NN_4.csv', index_col = 0)
y_train = pd.read_csv ('y_train_4.csv', index_col = 0)

y_train = y_train.prediabetes_bin_y
```

Note that the train data containing colinear variables is loaded here as KNN classifier is not inherently sensitive to colinearity.

## Optimise Hyperparameters for KNN Classifier

The hyperparameters for KNN classifiers include:
- number of neighbours (minimum number of neighbours required to be split as a separate group)
- weights for each distance
- how to measure the distance between points  
  
Since the KNN Classifier relies on distances between datapoints, the data will have to be scaled. As determined previously, min-max scaling is the most appropriate and performs best for the problem at hand.

### First run for hyperparameters

```python
# Define the pipeline
pipeline = Pipeline([
    ('scaler', MinMaxScaler()),  # Placeholder for scaler
    ('knn', KNeighborsClassifier())  # KNN classifier
])

# Define the parameter grid for grid search
param_grid = {
    'scaler': [MinMaxScaler()],  # Scaler options
    'knn__n_neighbors': [3, 555, 22221],  # Number of neighbors
    'knn__weights': ['uniform', 'distance'],  # Weight function
    'knn__metric' : ['manhattan', 'minkowski'], # Different distance measures for linear/binary
}

# Create the GridSearchCV object
grid_search = GridSearchCV(pipeline, param_grid, cv=5, verbose = 2, n_jobs = 4)

# Fit the GridSearchCV object on the data
grid_search.fit(X_train, y_train)

# Print the best parameters and best score
print("Best Parameters: ", grid_search.best_params_)
print("Best Score: ", grid_search.best_score_)
```

Above are the hyperparameters that produced the best scores on average for 5-fold cross-validation. These hyperparameters will be used as anchor for defining values for a second round of hyperparameter optimisation.

### Refine hyperparameters (1)

```python
# Define the parameter grid for grid search
param_grid = {
    'scaler': [MinMaxScaler()],  # Scaler options
    'knn__n_neighbors': [221, 443, 1001],  # Number of neighbors centered around 555
    'knn__weights': ['uniform'],  # Weight function
    'knn__metric' : ['minkowski'] # Different distance measures for linear/binary
}

# Create the GridSearchCV object
grid_search = GridSearchCV(pipeline, param_grid, cv=5, verbose = 2, n_jobs = 2)

# Fit the GridSearchCV object on the data
grid_search.fit(X_train, y_train)

# Print the best parameters and best score
print("Best Parameters: ", grid_search.best_params_)
print("Best Score: ", grid_search.best_score_)
```

### Refine hyperparameters (2)

```python
# Define the parameter grid for grid search
param_grid = {
    'scaler': [MinMaxScaler()],  # Scaler options
    'knn__n_neighbors': [21, 55, 111, 221, 333],  # Number of neighbors centered around 555
    'knn__weights': ['uniform'],  # Weight function
    'knn__metric' : ['minkowski'] # Different distance measures for linear/binary
}

# Create the GridSearchCV object
grid_search = GridSearchCV(pipeline, param_grid, cv=5, verbose = 2, n_jobs = 4)

# Fit the GridSearchCV object on the data
grid_search.fit(X_train, y_train)

# Print the best parameters and best score
print("Best Parameters: ", grid_search.best_params_)
print("Best Score: ", grid_search.best_score_)
```

The best scores are displayed above for a 5-fold cross-validation on the training dataset. These hyperparameters will now be used to build and train a KNN classifier on the entire training dataset and assess its performance on the test dataset.

## Build a KNN Classifier and Assess Performance

### Build KNN classifier

```python
# instantiate KNN classifier and pass best parameters
knn_class = KNeighborsClassifier(metric = 'minkowski', 
                                 n_neighbors = 111, 
                                 weights='uniform')
```

```python
# scale training data:
scaler = MinMaxScaler()
Xtrain_scaled = scaler.fit_transform (X_train)
```

```python
# Fit KNN classifier on scaled train data
knn_class.fit (Xtrain_scaled, y_train)
```

### Assess performance

#### Train set

```python
# for training dataset
# Get the predicted probabilities for the positive class
y_pred_proba = knn_class.predict_proba(Xtrain_scaled)[:, 1]
```

```python
y_pred = knn_class.predict (Xtrain_scaled)
# metrics
accuracy = accuracy_score(y_train, y_pred)
precision = precision_score(y_train, y_pred)
recall = recall_score(y_train, y_pred)
f1 = f1_score(y_train, y_pred)
# Display scores
print("Accuracy (training):", accuracy)
print("Precision (training):", precision)
print("Recall (training):", recall)
print("F1 Score (training):", f1)
```

#### Test set

```python
# load test data and scale
# Load test data:
X_test = pd.read_csv ('X_test_NN_4.csv', index_col = 0)
y_test = pd.read_csv ('y_test_4.csv', index_col = 0)

y_test = y_test.prediabetes_bin_y
# Scale
Xtest_scaled = scaler.transform (X_test)
```

```python
y_pred_proba = knn_class.predict_proba(Xtest_scaled)[:, 1]
```

```python
y_pred = knn_class.predict (Xtest_scaled)
# metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
# Display scores
print("Accuracy (training):", accuracy)
print("Precision (training):", precision)
print("Recall (training):", recall)
print("F1 Score (training):", f1)
```

Overall, KNN classfication performs very well, but does not outperform decision tree classifier. Therefore the winner is Decision Tree Classifier. Nevertheless some more metrics for KNN classifier performance will be displayed below.

### ROC and AUC

```python
# Calculate the false positive rate, true positive rate, and thresholds
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)

# Calculate the AUC score
auc = roc_auc_score(y_test, y_pred_proba)
```

```python
# Plot the ROC curve
plt.plot(fpr, tpr, label='ROC Curve (AUC = {:.2f})'.format(auc))
plt.plot([0, 1], [0, 1], 'k--', label='Random')
plt.text(0.6, 0.3, ('AUC = ' + str(round(auc, 4))), fontsize = 16)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend()
plt.show()
```

### Confusion matrix

```python
# Create confusion matrix
cm = confusion_matrix(y_test, y_pred)

class_names = ['non-prediabetes', 'prediabetes']

# Plot confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.xticks(np.arange(len(class_names))+0.5, class_names, rotation=45, fontsize = 12)
plt.yticks(np.arange(len(class_names))+0.5, class_names, rotation=0, fontsize = 12)
plt.xlabel('Predicted', fontsize = 14)
plt.ylabel('Actual', fontsize = 14)
plt.title('Confusion Matrix', fontsize = 16)
plt.show()
```

We can see we have a total of 3,663 false negatives and 14,479 false positives out of a total of 247,292 patients total in the test set.

### Classification report

```python
# Create classification report
report = classification_report(y_test, y_pred, target_names=class_names, output_dict=True)

# Plot classification report
sns.heatmap(pd.DataFrame(report).iloc[:-1, :].T, annot=True, cmap='Blues')
plt.xlabel('Metrics', fontsize = 14)
plt.ylabel('Classes', fontsize = 14)
plt.title('Classification Report', fontsize = 16)
plt.show()
```

### Precision-Recall Curve

```python
# Compute precision, recall, and threshold values
precision, recall, thresholds = precision_recall_curve(y_test, y_pred_proba)

# Calculate the average precision score
avg_precision = average_precision_score(y_test, y_pred_proba)

# Plot precision-recall curve
plt.plot(recall, precision, marker='.', label='KNN')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.legend(loc='lower left')
plt.grid(True)
plt.show()

# Print average precision score
print("Average Precision Score:", avg_precision)
```

The ideal curve would hug the top right corner indicating high precision and recall. Our model gives us some wiggle room between 0.75 and close to 1 for recall to capitalise on the trade-off between the to metrics by adjusting the threshold probability for calling a positive-class outcome (prediabetes). This would depend on the task at hand - whether we want to make sure that our predictions are accurate (more precision) or that we focus more on being able to capture all positive events (recall) even though our test produces more false positives.
